```python
import requests

url = "http://127.0.0.1:8000/upload"

files = [
    ("files", ("case.csv", open("files/case.csv", "rb"))),
    ("files", ("example.csv", open("files/example.csv", "rb"))),
    ("files", ("example_out.txt", open("files/example_out.txt", "rb"))),
    ("files", ("GANNDALF-onto.ttl", open("files/GANNDALF-onto.ttl", "rb")))
]

response = requests.post(url, files=files)

print("Status:", response.status_code)
print("Response:", response.json())
```
